```python
import random
from datetime import datetime
import time
import os
import csv
from gtts import gTTS
from googletrans import Translator
from IPython.display import Audio, display, HTML
import ipywidgets as widgets
from PIL import Image
import io

# Create directories
os.makedirs("excuse_logs", exist_ok=True)
log_file_path = os.path.join("excuse_logs", "daily_excuses.csv")

if not os.path.exists(log_file_path):
    with open(log_file_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["EmployeeID", "DateTime", "Context", "Mood", "Excuse", "Proof"])

# Basic CSS Styles
style = """
    <style>
        .excuse-box {
            border: 2px solid #4CAF50;
            border-radius: 10px;
            padding: 15px;
            margin-top: 10px;
            background-color: #f9f9f9;
            font-family: Arial, sans-serif;
            color: #333;
            animation: fadeIn 2s;
        }
        @keyframes fadeIn {
            from {opacity: 0;}
            to {opacity: 1;}
        }
        .excuse-header {
            color: #4CAF50;
            font-size: 20px;
            font-weight: bold;
        }
    </style>
"""
display(HTML(style))

# Contexts and emojis
contexts = {
    "work": [
        "üßë‚Äçüíª I'm caught up in a critical client meeting and will need a bit more time.",
        "üíª My system just crashed and I lost all progress, trying to recover it now.",
        "üåê There was a sudden internet outage in my area, I‚Äôll be back online shortly."
    ],
    "school": [
        "ü§í I was feeling unwell after lunch and couldn‚Äôt focus on the assignment.",
        "üìÑ My assignment file got corrupted right before submission‚Äîworking on a recovery.",
        "‚ö° There was a power cut and I couldn‚Äôt complete the work on time."
    ],
    "social": [
        "üöó I got stuck in unexpected traffic, really sorry for the delay.",
        "üë®‚Äçüë©‚Äçüëß‚Äçüë¶ There‚Äôs a family emergency I need to attend to‚Äîwill catch up later.",
        "ü§ß I‚Äôm not feeling well and don‚Äôt want to risk spreading anything."
    ],
     "family": [
        "üÜò I had to help a relative with a sudden health issue.",
        "üìû There was an urgent family discussion I couldn‚Äôt avoid.",
        "üôè I was asked to stay back for a family ritual."
    ],
    "pet": [
        "üê∂ My dog needed emergency care and I had to rush to the vet.",
        "üê± My cat went missing and I had to look for it around the neighborhood.",
        "üè• I had to rush my pet to the vet due to sudden illness."
    ],
    "weather": [
        "üåßÔ∏è Heavy rain flooded my neighborhood and I couldn‚Äôt leave the house.",
        "‚õàÔ∏è Storm warnings kept me from traveling for safety reasons.",
        "‚ö° My area experienced a sudden power outage due to lightning."
    ],
    "health": [
        "ü§ï I had a severe headache and couldn‚Äôt concentrate on anything.",
        "üå°Ô∏è I had a sudden fever and had to rest as per doctor‚Äôs advice.",
        "ü§¢ I experienced food poisoning symptoms unexpectedly and had to lie down."
    ],
    "travel": [
        "‚úàÔ∏è My flight was delayed due to bad weather conditions.",
        "üöå My bus broke down on the way and I had to wait for assistance.",
        "üöß There was a strike that disrupted transportation in my area."
    ],
    "technical": [
        "üíª My laptop crashed during an update and I lost some progress.",
        "üì° The Wi-Fi router stopped working, waiting for it to reconnect.",
        "üñ•Ô∏è The server was down and I couldn't access the system remotely."
    ],
    "mental health": [
        "üßò I needed to take a break for mental well-being and avoid burnout.",
        "üòû I felt overwhelmed and took a moment to calm down before continuing.",
        "üò∞ I had an anxiety episode and had to step away from everything."
    ],
    "guest": [
        "üö™ Unexpected guests arrived and I had to host them temporarily.",
        "üë®‚Äçüë©‚Äçüëß‚Äçüë¶ Relatives visited without prior notice and I couldn‚Äôt excuse myself.",
        "üëã A family friend dropped in urgently and I had to entertain them."
    ],
    "event": [
        "üôè I had to attend a religious ceremony at short notice with family.",
        "üöó A local event blocked my commute and I was stuck in traffic.",
        "üéâ I was volunteering at a community function and lost track of time."
    ],
    "hospital": [
        "üè• I had a minor surgery scheduled suddenly and couldn't inform earlier.",
        "ü©∫ A medical emergency required me to stay at the hospital longer than expected.",
        "üß¨ Hospital admission was necessary due to a sudden allergic reaction."
    ],
    "public transport": [
        "üöá The metro service was temporarily suspended causing huge delays.",
        "üöå My bus broke down midway and I had to wait for an alternate.",
        "üö¶ There was a major traffic jam due to a roadblock on my route."
    ],
    "technology failure": [
        "üîã My phone battery died while trying to connect online.",
        "üõú The company VPN was down, locking me out of the work system.",
        "‚òÅÔ∏è Cloud storage failed to sync my latest file before submission."
    ],
    "family emergency": [
        "üö® A close family member had a sudden medical situation I needed to attend.",
        "üì± There was an urgent family call and I had to step away immediately.",
        "üè† A domestic emergency required my attention and delayed my work."
    ],
    "natural disaster": [
        "üåä There was a flash flood warning issued for my area.",
        "üåç A small earthquake hit my locality and caused disruption.",
        "üå™Ô∏è High winds brought down trees, blocking roads and internet lines."
    ],
    "unexpected guests": [
        "üè° Neighbors showed up unexpectedly needing urgent help.",
        "üéä A surprise family gathering kept me occupied longer than planned.",
        "üë®‚Äçüë©‚Äçüë¶ Close relatives visited without any prior notice today."
    ],
    "work overload": [
        "üìù I got multiple urgent tasks dumped at the last moment.",
        "üìä I was assigned additional responsibilities and couldn‚Äôt complete everything in time.",
        "üìÖ Back-to-back meetings stretched longer than anticipated."
    ],
    "lost important items": [
        "üîë I lost my house keys and had to arrange emergency locksmith help.",
        "üÜî I misplaced my office ID badge and was delayed entering the building.",
        "üëõ My wallet was missing and I had to report it at the station."
    ],
    "power outage": [
        "üí° A transformer burst near my area causing prolonged blackouts.",
        "üîå Power fluctuation damaged my modem and delayed online work.",
        "‚ö° Unscheduled maintenance caused sudden electricity cut-off."
    ],
    "online meetings": [
        "üìπ Zoom crashed due to a system update right before my call.",
        "üé§ My mic wasn‚Äôt working and I had to troubleshoot during the meeting.",
        "üì∑ My camera stopped functioning suddenly during a critical session."
    ],
    "festival season": [
        "üéÜ Festival traffic and celebrations delayed my return.",
        "üéâ I had to participate in unexpected family rituals.",
        "üïØÔ∏è Festival-related electricity outage affected my area."
    ],
    "mental health break": [
        "üõå I needed a digital detox break to recharge mentally.",
        "üßò‚Äç‚ôÇÔ∏è I took a meditation break to regain focus.",
        "üòì I had to take some rest due to severe anxiety spikes."
    ],
    "vehicle breakdown": [
        "üö¥ My bike's chain snapped and I had to push it home.",
        "üöó Car tire burst suddenly while driving to the location.",
        "üõ†Ô∏è Engine trouble stranded me on the way unexpectedly."
    ],
    "oversleeping": [
        "‚è∞ My alarm didn‚Äôt ring due to a phone software bug.",
        "üò¥ I accidentally turned off the alarm and overslept.",
        "üõèÔ∏è I stayed up late fixing issues and missed the morning alarm."
    ]
}

# Proof templates
proof_templates = [
    "Email screenshot indicating delay: '{}'",
    "Chat log snippet: '{} - seen at 4:47 PM'",
    "GPS log: user last seen at different location when '{}'",
    "Medical note auto-generated for: '{}'",
    "Simulated calendar event overlap: '{}'"
]

# Cooldown settings
last_generated_time = 0
cooldown_seconds = 60  # 1 minute cooldown between excuses

# Functions
def generate_excuse(context, mood="Neutral"):
    excuses = contexts.get(context.lower(), [])
    if not excuses:
        return "No excuses available for this context."
    return random.choice(excuses)

def generate_proof(excuse):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    proof_type = random.choice(proof_templates).format(excuse)
    return f"[Proof - {timestamp}] {proof_type}"

def log_excuse(employee_id, context, mood, excuse, proof):
    with open(log_file_path, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([employee_id, datetime.now().strftime("%Y-%m-%d %H:%M:%S"), context, mood, excuse, proof])

def speak_text(text, lang='en'):
    tts = gTTS(text=text, lang=lang)
    tts.save("excuse.mp3")
    display(Audio("excuse.mp3", autoplay=True))

def multi_language_support(excuse, language_code):
    translator = Translator()
    try:
        translated = translator.translate(excuse, dest=language_code)
        return f"[{translated.dest.upper()}] {translated.text}"
    except Exception as e:
        return f"Translation error: {str(e)}"

def upload_image():
    upload_widget = widgets.FileUpload(accept='image/*', multiple=False)

    def on_upload_change(change):
        for filename, file_info in upload_widget.value.items():
            image_data = file_info['content']
            image = Image.open(io.BytesIO(image_data))
            display(image)
            with open(os.path.join("excuse_logs", filename), "wb") as f:
                f.write(image_data)
            display(HTML(f"<div class='excuse-box'>‚úÖ Image '{filename}' uploaded and saved.</div>"))

    upload_widget.observe(on_upload_change, names='value')
    display(upload_widget)

# User Inputs
employee_id_input = widgets.Text(
    placeholder='Enter Employee ID',
    description='Employee ID:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)

context_dropdown = widgets.Dropdown(
    options=list(contexts.keys()),
    description='Context:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)

mood_dropdown = widgets.Dropdown(
    options=["Neutral", "Sad", "Stressed", "Angry", "Excited"],
    description='Mood:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)

def on_context_selected(change):
    global last_generated_time
    current_time = time.time()

    if current_time - last_generated_time < cooldown_seconds:
        wait_time = int(cooldown_seconds - (current_time - last_generated_time))
        display(HTML(f"<div class='excuse-box' style='background-color:#FFCDD2;'>‚è≥ Please wait {wait_time}s before generating another excuse!</div>"))
        return

    last_generated_time = current_time

    context = context_dropdown.value
    mood = mood_dropdown.value
    employee_id = employee_id_input.value or "Unknown"

    excuse = generate_excuse(context, mood)
    proof = generate_proof(excuse)
    score = random.randint(70, 99)  # Believability Score

    log_excuse(employee_id, context, mood, excuse, proof)
    apology = f"Sincere apologies. Mood noted: {mood}. {excuse}"

    # Dynamic pastel colors
    pastel_colors = ["#FFFBF0", "#E0F7FA", "#FFF3E0", "#F3E5F5", "#E8F5E9", "#FFF8E1"]
    random_color = random.choice(pastel_colors)

    emoji = random.choice(contexts[context])

    display(HTML(f"""
        <div class='excuse-box' style='background-color: {random_color};'>
            <div class='excuse-header'>{emoji} Generated Excuse:</div>
            {excuse}<br><br>
            <div class='excuse-header'>üìú Proof:</div>
            {proof}<br><br>
            <div class='excuse-header'>üôè Apology:</div>
            {apology}<br><br>
            <div><b>üîÆ Believability Score:</b> {score}%</div>
        </div>
    """))

    speak_text(excuse)

    lang_dropdown = widgets.Dropdown(
        options={"English": "en", "Tamil": "ta", "Kannada": "kn", "Malayalam": "ml", "Hindi": "hi", "Japanese": "ja", "German": "de"},
        description='Translate To:',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='50%')
    )

    def on_language_selected(l_change):
        lang_code = l_change['new']
        translated_excuse = multi_language_support(excuse, lang_code)
        display(HTML(f"<div class='excuse-box' style='background-color: {random_color};'><b>üåç Translated Excuse:</b><br>{translated_excuse}</div>"))
        if lang_code != "en":
            speak_text(translated_excuse.split("] ", 1)[1], lang=lang_code)

    lang_dropdown.observe(on_language_selected, names='value')
    display(lang_dropdown)

    display(HTML("<div class='excuse-box'><b>üì∑ Upload an Image as Proof:</b></div>"))
    upload_image()

# Display Input Widgets
display(employee_id_input)
display(mood_dropdown)
context_dropdown.observe(on_context_selected, names='value')
display(context_dropdown)
```
